```python
numeric_df = df.select_dtypes(include=['number'])
corr = numeric_df.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot = True, cmap = 'rocket')
plt.title("Correlation Heatmap")
plt.savefig('Corelation Heatmap.png', bbox_inches='tight', dpi=300)
plt.show()
```

# **Key Insights from the Correlation Matrix**

**Government Expenditure and Enrollment:**

A positive correlation between gov_exp_pct_gdp and enrollment rates (school_enrol_primary_pct, school_enrol_secondary_pct, school_enrol_tertiary_pct) suggests that higher government spending on education is associated with higher enrollment rates.

**Literacy Rates and Enrollment:**

A strong positive correlation between lit_rate_adult_pct and enrollment rates indicates that higher literacy rates are associated with higher school enrollment.
**Pupil-Teacher Ratios:**

Negative correlations between pupil_teacher_primary/pupil_teacher_secondary and enrollment/literacy rates suggest that lower pupil-teacher ratios (more teachers per student) are associated with better education outcomes.

**Primary and Secondary Enrollment:**

A strong positive correlation between school_enrol_primary_pct and school_enrol_secondary_pct indicates that countries with high primary enrollment rates also tend to have high secondary enrollment rates.

```python
import plotly.graph_objects as go

# Create a base world map
fig = go.Figure()

# Add a scattergeo trace for the globe
fig.add_trace(go.Scattergeo(
    locationmode = 'country names',
    locations = countries,
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'red',
        symbol = 'x'
    ),
    name = 'Countries in Dataset'
))

# Update layout for a 3D globe view
fig.update_layout(
    title = 'World Education Data - Countries in Dataset',
    geo = dict(
        projection_type = 'orthographic',
        showland = True,
        showocean = True
    ),)
fig.show()
```
